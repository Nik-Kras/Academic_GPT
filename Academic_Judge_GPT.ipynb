````python
def transform_dataset(ddf):

    df = ddf.copy()

    # First, we convert the string representations of dictionaries into actual dictionaries
    for column in df.columns:
        if 'model' in column:
            df[column] = df[column].apply(eval)

    # Now, we expand these dictionaries into sub-columns
    df_expanded = pd.json_normalize(df.drop(columns=['Unnamed: 0', 'type']).to_dict('records'))

    # Create a new DataFrame with multi-level columns for better organization
    multi_col_df = pd.concat([df[['Unnamed: 0', 'type']], df_expanded], axis=1)
    multi_col_df.columns = pd.MultiIndex.from_tuples([tuple(c.split('.')) if '.' in c else (c, '') for c in multi_col_df.columns])

    return multi_col_df

new_df = transform_dataset(df)
new_df
````

To seect the best model with high accuracy I need to measure an 'overlap' of distributions. That would tell me how frequently this model is confused whether the category is good or bad for the given text.

The only idea currently that comes to my mind is: calculate mean and std for each distribution, transform them to gaussian distributions and calculate an overlapping area of two gaussians.

However, the issue is that I have outliers and my data isn't normally distributed. That is why I could use Wilcoxon rank sum test and select the model with the least p-value.

GPT:

'''

The Wilcoxon Rank-Sum Test, which is also known as the Mann-Whitney U test when applied to two independent samples, is a non-parametric test that is used to infer whether there is a statistically significant difference between the distributions of two independent samples. It is especially useful when the sample sizes are small and the data do not necessarily follow a normal distribution, which makes it more robust than t-tests in such scenarios.

Test Statistic: The test statistic is the sum of the ranks of the observations from the first sample. If this value is significantly high or low, it indicates that one of the samples consistently has higher or lower ranks compared to the other, suggesting a difference in distributions.

p-value: The p-value tells us about the probability of observing the data, or something more extreme, if the null hypothesis were true.
- If the p-value is less than the chosen significance level (typically 0.05), you reject the null hypothesis and conclude that there is a significant difference between the two groups.
- If the p-value is greater than the significance level, you fail to reject the null hypothesis, which suggests that there is not enough evidence to say that the distributions are different.

'''

````python
def evaluate_best_models_per_feature(df):

  features = {t for t in df.columns.get_level_values(1) if t != ""}
  model_names = sorted(list(set([c for c in set(df.columns.get_level_values(0)) if c.startswith("model")])))

  bad_text = df[df["type"] == "bad"]
  good_text = df[df["type"] == "good"]

  evaluation = {}

  for feature in features:
    evaluation[feature] = []
    bad_model = bad_text.xs(feature, axis=1, level=1, drop_level=False)
    good_model = good_text.xs(feature, axis=1, level=1, drop_level=False)

    for model in model_names:
      evaluation[feature].append(ranksums(bad_model[model], good_model[model]).pvalue[0])

  return pd.DataFrame.from_dict(evaluation, orient='index', columns=model_names)

evaluate_best_models_per_feature(new_df)
````

# 8 Statistical inference for best models per criteria

Now, for given best models I can simply calculate KDE (Kernel Density Estimation)

With this PDFs (Probability Density Functions) we can build a non-parametric statistical model to evaluate the probabilities of a new sample to be a bad or good paragraph

````python
def get_best_models_per_feature(df):
  model_evaluation = evaluate_best_models_per_feature(df)
  res = {}
  for key, eval_list in model_evaluation.iterrows():
    res[key] = str(eval_list.idxmin())
  res = pd.Series(res)
  return res


def visualise_one_kde(good_values, bad_values, title):

  import matplotlib.pyplot as plt
  from scipy.stats import gaussian_kde
  import numpy as np

  # Fit KDE to data
  good_kde = gaussian_kde(good_values)
  bad_kde = gaussian_kde(bad_values)

  # Values over which we'll calculate the KDE
  x_min = min(min(good_values), min(bad_values)) - 10
  x_max = max(max(good_values), max(bad_values)) + 10
  x_values = np.linspace(x_min, x_max, 1000)

  # Evaluate the KDE here
  good_kde_values = good_kde(x_values)
  bad_kde_values = bad_kde(x_values)

  # Plotting the KDEs
  plt.figure(figsize=(10, 6))
  plt.title(title)
  plt.plot(x_values, good_kde_values, label='Good KDE', color='blue')
  plt.plot(x_values, bad_kde_values, label='Bad KDE', color='red')
  plt.scatter(good_values, np.zeros_like(good_values), color='blue', zorder=5, label='Good Data Points')
  plt.scatter(bad_values, np.zeros_like(bad_values), color='red', zorder=5, label='Bad Data Points')
  plt.legend()
  plt.grid(True)
  plt.show()


def visualise_all_kde(df):

  res = get_best_models_per_feature(df)

  bad_text = df[df["type"] == "bad"]
  good_text = df[df["type"] == "good"]

  for feature, model in res.items():
    good_values = good_text[model][feature].to_list()
    bad_values = bad_text[model][feature].to_list()
    title = "{} for feature {}".format(model, feature)
    visualise_one_kde(good_values, bad_values, title)


visualise_all_kde(new_df)
````

````python
import json
import numpy as np

def generate_paragraph_metrics(paragraph, feature_model_map):
  """
  <PARAGRAPH> ->
  {"model_1": ..., "model_3": ...} ->
  {"feature_1": ..., "feature_2": ...}

  """
  only_mandatory_models = set(feature_model_map)
  model_index = [int(c[-1]) for c in only_mandatory_models]

  model_output = {}
  for ind in model_index:
    model_output[f"model_{ind}"] = json.loads(model_ask(paragraph, PROMPTS[ind]).content)

  print("General evaluation by every model")
  print(pd.DataFrame(model_output))

  output = {}
  for feature, model in feature_model_map.items():
    output[feature] = model_output[model][feature]

  print("Left only values from most useful models")
  print(pd.Series(output))

  return output


def predict_paragraph(paragraph, feature_model_map, statistical_model):
    # paragraph_model_evaluation is a dict, where each criteria is assigned with a number by the best model beforeahead
    # feature_model_map shows what model should we rely on while analysing specific feature
    # Statistical_model keeps KDE params for label estimation

    paragraph_metrics = generate_paragraph_metrics(paragraph, feature_model_map)

    output = {}
    for feature, kdes in statistica_model.items():
      model_to_use = feature_model_map[feature]

      # Calculate the probability density of the new data point in each distribution
      good_prob_density = kdes["good"].evaluate(paragraph_metrics[feature])[0]
      bad_prob_density = kdes["bad"].evaluate(paragraph_metrics[feature])[0]

      # Normalize the probabilities so they sum to 1
      total_density = good_prob_density + bad_prob_density
      prob_good = good_prob_density / total_density if total_density > 0 else 0
      prob_bad = bad_prob_density / total_density if total_density > 0 else 0

      output[feature] = {
        "good": prob_good,
        "bad": prob_bad
      }

    return output
````
